```python
# Dependencias:
import requests
from bs4 import BeautifulSoup
```

```python
url_equipos = 'https://www.marca.com/baloncesto/acb/equipos.html?intcmp=MENUMIGA&s_kw=equipos-y-jugadores'
```

```python
## Entidades.

# Definición de equipo.
class acb_equipo():
    def __init__(self, nombreequipo, entrenador, jugadores):
        self._nombreequipo = nombreequipo
        self._entrenador = entrenador
        self._jugadores = jugadores
        
    def entrenador(self):
        return _entrenador
    
    def jugadores(self):
        return _jugadores
    
    def printInfoEquipo(self):
        print('Equipo: ', self._nombreequipo)
        print('Equipo: ', self._entrenador)
        for j in self._jugadores:
            j.printInfoJugador()
    
# Definición de jugador
class acb_jugador():
    def __init__(self, dorsal, nombre, fechanacimiento, posicion, altura, peso):
        self._dorsal = dorsal
        self._nombre = nombre
        self._fechanacimiento=fechanacimiento
        self._posicion=posicion
        self._altura =altura
        self._peso = peso   
    
    def printInfoJugador(self):
        print('Dorsal: ', self._dorsal)
        print('nombre: ', self._nombre)
        print('fechanacimiento: ', self._fechanacimiento)
        print('posicion: ', self._posicion)
        print('altura: ', self._altura)
        print('peso: ', self._peso)
```

```python
def recuperarPagina(url):
    return requests.get(url)

def recuperarJugadores(html_equipo):
    acb_jugadores = []
    html_jugadores = html_equipo.select ('.plantilla tr')    
    for jugador in html_jugadores:
        acb_jugadores.append(acb_jugador(
                jugador.select('.dorsal')[0].text,
                jugador.select('.nombre')[0].text,
                jugador.select('.fecha-nacimiento')[0].text,
                jugador.select('.posicion')[0].text,
                jugador.select('.altura')[0].text,
                jugador.select('.peso')[0].text))        
    return acb_jugadores

def recuperarPlantillas(soup):    
    acb_equipos = []
    
    html_equipos = soup.select('div .plantillas-equipos li')
    for html_equipo in html_equipos:
        nombreequipo = html_equipo.select('h2')[0].text
        nombreentrenador = html_equipo.select('dd')[0].text    
        juagadores = recuperarJugadores(html_equipo)        
        acb_equipos.append(acb_equipo(nombreequipo, nombreentrenador, juagadores)) 
        
    return acb_equipos
```

```python
page = recuperarPagina(url_equipos)
soup = BeautifulSoup(page.content, 'html.parser')

plantillas = recuperarPlantillas (soup)
```
